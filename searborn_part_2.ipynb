```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
tips = pd.read_csv("./data/tips.csv")
iris = sns.load_dataset("iris")
```

### Categorical plot 

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# strip plot

# axes level function
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
)
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x="day", y="total_bill", kind="strip")
```

```python
# jitter
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", jitter=0.2, hue="sex")
```

```python
# swarmplot
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
# swarmplot
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex")
```

### Boxplot

```python
# boxplot
sns.boxplot(data=tips, x="day", y="total_bill")
```

```python
# using catplot
sns.catplot(data=tips, x="sex", y="total_bill", kind="box")
```

```python
# hue
# boxplot
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
# single boxplot -> numerical col
sns.boxenplot(data=tips, y="total_bill")
```

### Violinplot (boxplot+kdeplot)

```python
# Violinplot

sns.violinplot(data=tips, x="day", y="total_bill")
```

```python
# hue

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex")
```
