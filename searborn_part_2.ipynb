```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
```

```python
tips = pd.read_csv("./data/tips.csv")
iris = sns.load_dataset("iris")
```

### Categorical plot 

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# strip plot

# axes level function
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
)
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x="day", y="total_bill", kind="strip")
```

```python
# jitter
sns.catplot(data=tips, x="day", y="total_bill", kind="strip", jitter=0.2, hue="sex")
```

```python
# swarmplot
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
# swarmplot
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex")
```

### Boxplot

```python
# boxplot
sns.boxplot(data=tips, x="day", y="total_bill")
```

```python
# using catplot
sns.catplot(data=tips, x="sex", y="total_bill", kind="box")
```

```python
# hue
# boxplot
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
# single boxplot -> numerical col
sns.boxenplot(data=tips, y="total_bill")
```

### Violinplot (boxplot+kdeplot)

```python
# Violinplot

sns.violinplot(data=tips, x="day", y="total_bill")
```

```python
# hue

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
# hue

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
```

```python
# barplot

sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.min)
```

When there are multiple observations in each category , it also uses bootstrapping to compute a confidence interval around the estimate , which is plotted using error bars

```python
# barplot

sns.barplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.min)
```

```python
# point plot

sns.pointplot(data=tips, x="sex", y="total_bill", hue="smoker", estimator=np.min)
```

```python
sns.barplot(data=tips, x="sex", y="total_bill")
```

```python
sns.pointplot(data=tips, x="sex", y="total_bill", hue="smoker", errorbar=None)
```

A special case for the bar plot is when to show the number of observations in each category rather than computing a statistics for a second variable . This is similar to a histogram over a categorical , rather than quantitative variable .  

```python
sns.countplot(data=tips, x="sex", hue="day")
```

```python
# faceting using catplot

sns.catplot(data=tips, x="sex", y="total_bill", col="smoker", kind="box", row="time")
```

### Regression plot(Linear regration)
- reg plot 
- implot

In the simplest invocation , both function draw a scatterplot of two variables , x and y , and then fit the regression model y~x and plot the resulting regression line and a 95% confidence for that regression 

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# axes level
# hue parameter is not available
sns.regplot(
    data=tips,
    x="total_bill",
    y="tip",
)
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
)
```

```python
# residual plot
sns.residplot(data=tips, x="total_bill", y="tip")
```

A second way to plot facet plots -> facetplot


```python
sns.catplot(data=tips, x="sex", y="total_bill", kind="violin", col="day", row="time")
```

```python
g = sns.FacetGrid(data=tips, col="day", row="time")


g.map(sns.boxenplot, "sex", "total_bill")
```

```python
g = sns.FacetGrid(data=tips, col="day", row="time")


g.map(sns.violinplot, "sex", "total_bill")
```

Plotting Pairwise Relationship (PairGrid vs PairPlot)

```python
iris = sns.load_dataset("iris")
```

```python
sns.pairplot(iris, hue="species")
```

```python
# pair grid

g = sns.PairGrid(data=iris, hue="species")

g.map(
    sns.scatterplot,
)
```

```python
# map_diag

g = sns.PairGrid(data=iris, hue="species")

g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag

g = sns.PairGrid(data=iris, hue="species")

g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
# map_diag

g = sns.PairGrid(data=iris, hue="species")

g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data=iris, hue="species")

g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
